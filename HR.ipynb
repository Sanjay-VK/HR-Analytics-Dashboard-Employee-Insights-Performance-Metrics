```python
import pandas as pd

# Load the dataset
file_path = "C:/Users/Hp/Desktop/HR/HR Data.xlsx - HR data.csv"
df = pd.read_csv(file_path)

# Display initial dataset info
df.info()

# Drop unnecessary columns
drop_columns = ["emp no", "Employee Number", "Over18", "Standard Hours"]
df_cleaned = df.drop(columns=drop_columns, errors='ignore')
```

```python
df_cleaned.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="Attrition", data=df_cleaned, palette="Set2")
plt.title("Employee Attrition Count")
plt.xticks(ticks=[0, 1], labels=["No Attrition", "Attrition"])
plt.show()

```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df_cleaned["Age"], bins=20, kde=True, color="lightgreen")
plt.title("Age Distribution of Employees")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
df_cleaned.to_csv("HR_NEW.csv", index=False)
```
